```python
def update(x,lis,j,i):
    x = time.ctime(x).split(' ')
    if(x[i]==lis[j]):
        return 1
    else:
        return 0
```

```python
days = ['d1','d2','d3','d4','d5','d6','d7']
lis = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
j = 0
for i in days:
    df[i] = df['launched_at'].apply(lambda x:update(x,lis,j))
    j+=1
```

```python
df.head()
```

```python
mnths = ['m1','m2','m3','m4','m5','m6','m7','m8','m9','m10','m11','m12']
lis2 = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
j=0
for i in mnths:
    df[i] = df['launched_at'].apply(lambda x:update(x,lis2,j,1))
    j+=1
```

```python
def update2(x,i):
    x = time.ctime(x).split(' ')
    if(len(x[2])!=0):
        if(x[2]==i):
            return 1 
        else:
            return 0
    else:
        if(x[3]==i):
            return 1 
        else:
            return 0
```

```python
for i in range(1,32):
    df[str(i)] = df['launched_at'].apply(lambda x:update2(x,str(i)))
```
