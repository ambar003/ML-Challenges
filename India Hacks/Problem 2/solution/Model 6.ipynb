```python
xgb = XGBClassifier(learning_rate =0.01, n_estimators=1000, max_depth=6,
                    min_child_weight=7, gamma=0.01, subsample=0.8, colsample_bytree=0.8,
                    objective= 'binary:logistic', nthread=4, seed=2017)
```

# Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn import cross_validation,metrics
from sklearn.grid_search import GridSearchCV
```

```python
param_test = {
    'max_depth':[5,10,15,20,25],
    'min_samples_split':[2,10,20,50,100]
}
gsearch = GridSearchCV(estimator = RandomForestClassifier(n_estimators=100, max_depth=2,min_samples_split=2,
                                                          min_samples_leaf=1,n_jobs=-1,random_state=2017),
                        param_grid = param_test,n_jobs=4,iid=False, cv=10,scoring='roc_auc')
gsearch.fit(X_train,y)
gsearch.best_params_, gsearch.best_score_
```

```python
param_test = {
    'min_samples_leaf':[1,2,5,10,15]
}
gsearch = GridSearchCV(estimator = RandomForestClassifier(n_estimators=100, max_depth=15,min_samples_split=10,
                                                          min_samples_leaf=1,n_jobs=-1,random_state=2017),
                        param_grid = param_test,n_jobs=4,iid=False, cv=10,scoring='roc_auc')
gsearch.fit(X_train,y)
gsearch.best_params_, gsearch.best_score_
```

```python
rfr = RandomForestClassifier(n_estimators=1000, max_depth=15,min_samples_split=10,
                                                          min_samples_leaf=10,n_jobs=-1,random_state=2017)
```

```python
rfr.fit(X_train,y)
```

```python
pred = rfr.predict_proba(X_test[features])
```

```python
submission = pd.DataFrame(data=pred[:,1],index=test_id,columns=['segment'])
submission.head()
```

```python
submission.to_csv('rfr.csv')
```
