```python
from xgboost.sklearn import XGBClassifier
from sklearn.cross_validation import train_test_split
from sklearn.metrics import roc_auc_score
```

```python
titles = pd.read_csv('added_titles.csv')
titles.head()
```

```python
features = list(X.columns)
features.remove('segment')
```

```python
X_train = X[features]
y = X['segment']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_train, y, test_size=0.20, random_state=42)
```

```python
xgb = XGBClassifier()
xgb.fit(x_train,y_train)
pred = xgb.predict_proba(x_test)
score = roc_auc_score(y_test,pred[:,1])
print(score)
```

```python
col = list(titles.columns)
```

```python
max_score = score
max_score
```

```python
for i in col:
    x_train[i] = titles.loc[:,i]
    x_test[i] = titles.loc[:,i]
    xgb = XGBClassifier()
    xgb.fit(x_train,y_train)
    pred = xgb.predict_proba(x_test)
    score = roc_auc_score(y_test,pred[:,1])
    if(score>max_score):
        max_score = score
        print(i+" "+str(max_score))
    else:
        x_train = x_train.drop([i],axis=1)
        x_test = x_test.drop([i],axis=1)
```
