```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
test = pd.read_csv('test.csv')
test.head()
```

```python
df = pd.concat([train,test],axis=0)
df.head()
```

```python
dic = {}
```

```python
def update(x):
    if type(x) is str:
        s = x.split(',')
        for i in s:
            index = i.rfind(':')
            title_name = i[:index]
            count = dic.get(title_name,0)
            dic[title_name] = count+1
```

```python
df['titles'] = df['titles'].apply(lambda x:update(x))
```

```python
len(dic)
```

```python
lis = sorted(dic.items(),key=lambda x:(-x[1],x[0]))
```

```python
titles = []
total = 0
for i in range(200):
    titles.append(lis[i][0])
    total=total+lis[i][1]
```

```python
title_dic = {}
j=0
for i in titles:
    title_dic[i] = "t"+str(j)
    j+=1
```

```python
lis
```

```python
len(title_dic)
```

```python
newdf = pd.DataFrame(data=list(title_dic.items()),columns=['title','t'])
newdf.head()
```

```python
del [df]
```

```python
for i in titles:
    train[title_dic[i]]=0
    test[title_dic[i]]=0
```

```python
def update1(x):
    temp=[]
    if type(x) is str:
        s = x.split(',')
        for i in s:
            index = i.rfind(':')
            title_name = i[:index]
            temp.append(title_name)
    return temp    
```

```python
train['titles'] = train['titles'].apply(lambda x:update1(x))
test['titles'] = test['titles'].apply(lambda x:update1(x))
```

```python
train.head()
```

```python
def check(x,i):
    if i in x:
        return 1
    else:
        return 0
```

```python
for i in titles:
    t = title_dic[i]
    train[t] = train['titles'].apply(lambda x:check(x,i))
```
