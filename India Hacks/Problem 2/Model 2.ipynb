```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
def convert_to_list(x):
    s = x.split(',')
    temp = []
    for i in s:
        index = i.rfind(':')
        city_name = i[:index]
        temp.append(city_name)
    return temp    
```

```python
train['cities'] = train['cities'].apply(lambda x:convert_to_list(x))
```

```python
train['cities'].head()
```

```python
test = pd.read_csv('test.csv')
test.head()
```

```python
test['cities'] = test['cities'].apply(lambda x:convert_to_list(x))
test['cities'].head()
```

```python
def top_cities(df,dic):
    for i in range(len(df)):
        cities = df.iloc[i]['cities']
        for j in cities:
            x = dic.get(j,0)
            dic[j] = x+1
```

```python
df = pd.concat([train,test],axis=0)
df.head()
```

```python
dic= {}
top_cities(df,dic)
```

```python
lis = sorted(dic.items(),key=lambda x:(-x[1],x[0]))
```

```python
del [df]
```

```python
cities = []
total = 0
for i in range(500):
    cities.append(lis[i][0])
```

```python
cities
```

```python
for i in cities:
    train[i]=0
    test[i]=0
```

```python
train.head()
```

```python
def check(x,i):
    if i in x:
        return 1
    else:
        return 0
```

```python
for i in cities:
    train[i] = train['cities'].apply(lambda x:check(x,i))
```

```python
train.head()
```

```python
train.to_csv('added_cities.csv',index=False)
```

```python
for i in range(len(test)):
    city = test.loc[i,'cities']
    for j in city:
        if j in cities:
            test.loc[i,j]=1
```
