```python
import pandas as pd
import numpy as np
```

```python
% pylab inline
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
train = pd.read_csv('train.csv')
```

```python
train.info()
```

```python
train['DetectedCamera'].value_counts()
```

```python
train.hist()
```

```python
le = LabelEncoder()
```

```python
le.fit(train['DetectedCamera'])
```

```python
list(le.classes_)
```

```python
train['DetectedCamera'] = le.transform(train['DetectedCamera'])
```

```python
train['SignFacing (Target)'].value_counts()
```

```python
train['SignFacing (Target)'] = le.transform(train['SignFacing (Target)'])
```

```python
train['SignFacing (Target)'].value_counts()
```

```python
train['DetectedCamera'].value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn import cross_validation,metrics
from sklearn.grid_search import GridSearchCV
```

```python
RandomForestClassifier(n_estimators=10, criterion='gini', max_depth=None, min_samples_split=2, min_samples_leaf=1, 
                       max_features='auto',n_jobs=1, random_state=3)
```

```python
train.columns
```

```python
features = ['DetectedCamera', 'AngleOfSign', 'SignAspectRatio', 'SignWidth',
       'SignHeight']
```

```python
X = train[features]
```

```python
y = train['SignFacing (Target)']
```

```python
param_test = {
    'max_depth':[5,10,15,20,25],
    'min_samples_split':[2,10,20,50,100]
}
gsearch = GridSearchCV(estimator = RandomForestClassifier(n_estimators=100, max_depth=2,min_samples_split=2,
                                                          min_samples_leaf=1,n_jobs=-1,random_state=111),
                        param_grid = param_test,n_jobs=4,iid=False, cv=10)
gsearch.fit(X,y)
gsearch.best_params_, gsearch.best_score_
```

```python
param_test = {
    'max_depth':[7,9,10,11,13],
    'min_samples_split':[40,50,60]
}
gsearch = GridSearchCV(estimator = RandomForestClassifier(n_estimators=100, max_depth=2,min_samples_split=2,
                                                          min_samples_leaf=1,n_jobs=-1,random_state=111),
                        param_grid = param_test,n_jobs=4,iid=False, cv=10)
gsearch.fit(X,y)
gsearch.best_params_, gsearch.best_score_
```

```python
param_test = {
    'min_samples_leaf':[1,2,5,10,15]
}
gsearch = GridSearchCV(estimator = RandomForestClassifier(n_estimators=100, max_depth=10,min_samples_split=50,
                                                          min_samples_leaf=1,n_jobs=-1,random_state=111),
                        param_grid = param_test,n_jobs=4,iid=False, cv=10)
gsearch.fit(X,y)
gsearch.best_params_, gsearch.best_score_
```

```python
param_test = {
    'min_samples_leaf':[7,9,10,13]
}
gsearch = GridSearchCV(estimator = RandomForestClassifier(n_estimators=100, max_depth=8,min_samples_split=3,
                                                          min_samples_leaf=1,n_jobs=-1,random_state=111),
                        param_grid = param_test,n_jobs=4,iid=False, cv=10)
gsearch.fit(X,y)
gsearch.best_params_, gsearch.best_score_
```

```python
param_test = {
    'n_estimators':[350,400,450]
}
gsearch = GridSearchCV(estimator = RandomForestClassifier(n_estimators=100, max_depth=10,min_samples_split=50,
                                                          min_samples_leaf=1,n_jobs=-1,random_state=111),
                        param_grid = param_test,n_jobs=4,iid=False, cv=10)
gsearch.fit(X,y)
gsearch.best_params_, gsearch.best_score_
```

```python
param_test = {
    'bootstrap':['True','False']
}
gsearch = GridSearchCV(estimator = RandomForestClassifier(n_estimators=1000, max_depth=10,min_samples_split=20,
                                                          min_samples_leaf=10,n_jobs=-1,random_state=111),
                        param_grid = param_test,n_jobs=4,iid=False, cv=10)
gsearch.fit(X,y)
gsearch.best_params_, gsearch.best_score_
```

```python
test = pd.read_csv('test.csv')
```

```python
test['DetectedCamera'] = le.transform(test['DetectedCamera'])
```

```python
prob = rf.predict_proba(test[features])
```

```python
submission = pd.DataFrame(data=prob,index = test.Id,columns=['Front','Left','Rear','Right'])
```
