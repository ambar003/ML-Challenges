```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
test = pd.read_csv('test.csv')
```

```python
test.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
s = 'X'+str(1)
s
```

```python
for i in range(9):
    if(i==7):
        continue
    le = LabelEncoder()
    s = 'X'+str(i)
    train[s] = le.fit_transform(train[s])
    test[s] = le.fit_transform(test[s])
```

```python
train.head()
```

```python
test.head()
```

```python
train.to_csv('updated_train.csv')
test.to_csv('updated_test.csv')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin = LinearRegression()
```

```python
features = list(train.columns)
```

```python
features.remove('ID')
```

```python
target = 'y'
```

```python
features.remove('y')
```

```python
lin.fit(train[features],train[target])
```

```python
pred = lin.predict(test[features])
```

```python
submission = pd.DataFrame(data = pred,index=test.ID,columns=['y'])
submission.head()
```

```python
submission.to_csv('linear_model.csv')
```
